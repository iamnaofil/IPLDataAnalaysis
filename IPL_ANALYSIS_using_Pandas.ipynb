**Data Analyst Make These as Functions to Retrive Top Winning Team in Given Season**

```python
#You can run this
def most_win_teams(df,year):
    """
    created by Momin Naofil
    takes Argument in Year
    Output top 5 winning team in that year
    """
    mask=df['season']==year
    winmask=df[mask]
    return winmask['winner'].value_counts().head(5)
year=int(input('Enter the season  :  '))
most_win_teams(df,year)
```

**Make Function out of this**

```python
def run_score(batsman_name):
    mask=delivery['batsman']==batsman_name
    vk=delivery[mask]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]

```

```python
run_score('DA Warner')
```

```python
run_score('V Kohli')
```

```python
run_score('G Gambhir')
```

```python
run_score('MS Dhoni')
```

```python
#Find the most destructive death over batsman in history of IPL
#On The Basis of Strike Rate = (Number of Run / Number of Balls)/100
#minimun 200 Balls in over 16-20(is another criteria)
```

```python
death_over=delivery[delivery['over']>15]
```

```python

allBatsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
#Runs Scored by these 43 batsman,
#Balls played by these 43 batsman
```

```python
#remeber we have created list of those 43 batsman so we use isin() function
delivery['batsman'].isin(batsman_list)
```
